```python
#연령별 생존율을 히스토그램으로 시각화하기
# 연령은 연속 값이므로 히스토그램을 만들어 확인한다.
# 히스토그램 -> .hist(), plt.hist()
# 여러 히스토그램이 동시에 보이게 하려면 -> histtype에 barstacked
# 히스토그램 빈 개수를 나타내는 bins
# 라벨을 지정하는 -> label
# 라벨을 지정하면 plt.legend() 를 써서 내부에 라벨을 표시할 수 도 있다.

plt.hist(x=[train_df.Age[train_df.Survived==0], train_df.Age[train_df.Survived==1]], bins=9, histtype='barstacked', label=['Death', 'Survived'])
plt.legend()
```

```python
# 카테고리 변수를 더미 변수화하기
# 상관관계의 계산이나 머신 러닝 등 처리는 수치 데이터에서만 실행할 수 있다. -> 원 핫 인코딩
# drop_first라고 하면 첫번째 열을 지울 수 있다. 
train_df_corr = pd.get_dummies(train_df, columns=['Sex'], drop_first=True)
train_df_corr = pd.get_dummies(train_df_corr, columns=['Embarked'])
```

```python
train_df_corr.head()
```

```python
# 상관 행렬(상관 계수를 나열하여 -1~1의 수치로 확인하는 것)
train_corr = train_df_corr.corr()
```

```python
train_corr
```

```python
# 히트맵으로 시각화하기
# 
plt.figure(figsize=(9,9))
sns.heatmap(train_corr, vmax=1, vmin=-1, center=0, annot=True)
```
